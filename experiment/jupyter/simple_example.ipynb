```python


import pandas as pd
import matplotlib.pylab as plt

df = pd.read_csv('../datasets/iris.csv')

setosa = df[df.Species=='Iris-setosa']
versicolor = df[df.Species=='Iris-versicolor']
virginica = df[df.Species=='Iris-virginica']

s_l = plt.plot(setosa['SepalLength'],setosa['PetalLength'],'bo')
vi_l = plt.plot(virginica['SepalLength'],virginica['PetalLength'],'go')
ve_l = plt.plot(versicolor['SepalLength'],versicolor['PetalLength'],'ro')
lx = ve_l[0].axes.set_xlabel('Sepal Length')
ly = ve_l[0].axes.set_ylabel('Petal Length')
plt.show()
```

```python
s_w = plt.plot(setosa['SepalWidth'],setosa['PetalWidth'],'bo')
vi_w = plt.plot(virginica['SepalWidth'],virginica['PetalWidth'],'go')
ve_w = plt.plot(versicolor['SepalWidth'],versicolor['PetalWidth'],'ro')
lx = ve_w[0].axes.set_xlabel('Sepal Width')
ly = ve_w[0].axes.set_ylabel('Petal Width')
plt.show()
```
